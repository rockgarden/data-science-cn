```python
# -*- coding: utf-8 -*-

import pandas as pd
import os
from datetime import datetime

# =============================================================================
# åŠ è½½ä¸¤ä¸ªæ•°æ®é›†
# =============================================================================

nls97 = pd.read_csv("data/nls97.csv").set_index("personid")

covidtotals = pd.read_csv("data/covidtotals.csv", parse_dates=['lastdate'])
covidtotals.set_index("iso_code", inplace=True)


def describe_dataset(df, name="æ•°æ®é›†", export_html=False, output_dir="reports"):
    """
    å¯¹ DataFrame è¿›è¡Œå…¨é¢æ¢æŸ¥ï¼Œå¹¶å¯é€‰å¯¼å‡ºä¸º HTML æŠ¥å‘Šã€‚

    å‚æ•°ï¼š
    df (pd.DataFrame): è¦åˆ†æçš„æ•°æ®æ¡†
    name (str): æ•°æ®é›†åç§°
    export_html (bool): æ˜¯å¦å¯¼å‡ºä¸º HTML
    output_dir (str): HTML æ–‡ä»¶ä¿å­˜ç›®å½•
    """
    # å±€éƒ¨æ˜¾ç¤ºè®¾ç½®ï¼ˆä»…å½±å“æ‰“å°è¾“å‡ºï¼‰
    with pd.option_context(
        'display.width', 70,
        'display.max_columns', 10,
        'display.max_rows', 100,
        'display.float_format', '{:,.0f}'.format
    ):
        print(f"{'='*60}")
        print(f"{name}")
        print(f"{'='*60}")

        # åŸºæœ¬ä¿¡æ¯
        shape = f"{df.shape[0]:,} è¡Œ Ã— {df.shape[1]} åˆ—"
        index_name = df.index.name or df.index.dtype
        unique_index = df.index.nunique()
        total_missing = df.isna().sum().sum()
        memory_usage = df.memory_usage(deep=True).sum() / 1e6
        dtypes_count = df.dtypes.value_counts()

        print(f"{'æ•°æ®å½¢çŠ¶':<20}: {shape}")
        print(f"{'ç´¢å¼•åç§°':<20}: {index_name}")
        print(f"{'å”¯ä¸€ç´¢å¼•æ•°é‡':<20}: {unique_index}")
        print(f"{'æ€»ç¼ºå¤±å€¼æ•°':<20}: {total_missing:,}")
        print(f"{'å†…å­˜å ç”¨':<20}: {memory_usage:.2f} MB")

        # æ•°æ®ç±»å‹ç»Ÿè®¡
        for dtype, count in dtypes_count.items():
            print(f"{'æ•°æ®ç±»å‹':<20}: {count} åˆ— {dtype}")

        # ç¼ºå¤±å€¼ Top 5
        missing_top5 = df.isna().sum().sort_values(ascending=False).head(5)
        print(f"\n{'='*60}")
        print("ç¼ºå¤±å€¼æœ€å¤šçš„å‰5ä¸ªå­—æ®µ")
        print(f"{'='*60}")
        if missing_top5.sum() > 0:
            for col, miss in missing_top5.items():
                print(f"{col:<25} : {miss:>8,} ({miss/len(df)*100:.1f}%)")
        else:
            print("æ²¡æœ‰ç¼ºå¤±å€¼ã€‚")

        # æ•°å€¼å‹å˜é‡ç»Ÿè®¡
        numeric_cols = df.select_dtypes(include='number').columns
        num_stats = None
        if len(numeric_cols) > 0:
            num_stats = df[numeric_cols].describe().round(0)
            num_missing = df[numeric_cols].isna().sum()
            num_stats.loc['ç¼ºå¤±æ•°é‡'] = num_missing
            print(f"\n{'='*60}")
            print("æ•°å€¼å‹å˜é‡ - æè¿°æ€§ç»Ÿè®¡ï¼ˆå‰5åˆ—ï¼‰")
            print(f"{'='*60}")
            print(num_stats.head().to_string())

        # åˆ†ç±»å˜é‡ç¤ºä¾‹
        cat_cols = df.select_dtypes(exclude='number').columns
        cat_example = {}
        if len(cat_cols) > 0:
            for col in cat_cols[:3]:
                top_val = df[col].value_counts(dropna=False).head(1)
                nan_count = df[col].isna().sum()
                cat_example[col] = {
                    'most': f"{list(top_val.index)[0]!r} ({top_val.iloc[0]:,})",
                    'missing': nan_count
                }
            print(f"\n{'='*60}")
            print("åˆ†ç±»å˜é‡ - æœ€å¸¸è§å€¼ï¼ˆå‰3ä¸ªï¼‰")
            print(f"{'='*60}")
            for col, info in cat_example.items():
                print(f"{col}: æœ€å¸¸è§={info['most']}, ç¼ºå¤±={info['missing']:,}")

        # æ ·æœ¬é¢„è§ˆ
        sample_data = df.sample(2, random_state=1) if len(df) >= 2 else df
        sample_transposed = sample_data.T

        print(f"\n{'='*60}")
        print("æ ·æœ¬æ•°æ®ï¼ˆè½¬ç½®ï¼‰")
        print(f"{'='*60}")
        print(sample_transposed)

        # ================================
        # å¯¼å‡ºä¸º HTML æŠ¥å‘Š
        # ================================
        if export_html:
            # åˆ›å»ºè¾“å‡ºç›®å½•
            os.makedirs(output_dir, exist_ok=True)
            timestamp = datetime.now().strftime("%Y%m%d_%H%M")
            filename = f"{output_dir}/{name.replace(' ', '_')}_{timestamp}.html"

            # æ„å»º HTML å†…å®¹
            html_template = f"""
            <!DOCTYPE html>
            <html lang="zh">
            <head>
                <meta charset="UTF-8">
                <title>{name} - æ•°æ®æ¢æŸ¥æŠ¥å‘Š</title>
                <style>
                    body {{ font-family: "Microsoft YaHei", sans-serif; margin: 40px; background: #f9f9f9; color: #333; }}
                    h1, h2 {{ color: #0056b3; }}
                    table {{ border-collapse: collapse; width: 100%; margin: 20px 0; background: white; box-shadow: 0 0 10px rgba(0,0,0,0.1); }}
                    th, td {{ padding: 12px; text-align: left; border: 1px solid #ddd; }}
                    th {{ background-color: #007acc; color: white; }}
                    tr:nth-child(even) {{ background-color: #f2f2f2; }}
                    .section {{ margin: 30px 0; }}
                    .info {{ line-height: 1.8; background: #eef7ff; padding: 15px; border-radius: 8px; }}
                </style>
            </head>
            <body>
                <h1>ğŸ“Š {name} - æ•°æ®æ¢æŸ¥æŠ¥å‘Š</h1>
                <p><strong>ç”Ÿæˆæ—¶é—´ï¼š</strong> {datetime.now().strftime('%Yå¹´%mæœˆ%dæ—¥ %H:%M')}</p>

                <div class="section">
                    <h2>ğŸ“Œ åŸºæœ¬ä¿¡æ¯</h2>
                    <div class="info">
                        <strong>æ•°æ®å½¢çŠ¶ï¼š</strong> {shape}<br>
                        <strong>ç´¢å¼•åç§°ï¼š</strong> {index_name}<br>
                        <strong>å”¯ä¸€ç´¢å¼•æ•°é‡ï¼š</strong> {unique_index:,}<br>
                        <strong>æ€»ç¼ºå¤±å€¼æ•°ï¼š</strong> {total_missing:,}<br>
                        <strong>å†…å­˜å ç”¨ï¼š</strong> {memory_usage:.2f} MB
                    </div>
                    <h3>æ•°æ®ç±»å‹åˆ†å¸ƒ</h3>
                    <ul>
            """

            for dtype, count in dtypes_count.items():
                html_template += f"<li>{count} åˆ— <code>{dtype}</code></li>\n"
            html_template += "</ul>"

            # ç¼ºå¤±å€¼ Top 5 è¡¨æ ¼
            html_template += """
                    <h3>ç¼ºå¤±å€¼æœ€å¤šçš„å‰5ä¸ªå­—æ®µ</h3>
                    <table>
                        <tr><th>å­—æ®µå</th><th>ç¼ºå¤±æ•°é‡</th><th>ç¼ºå¤±æ¯”ä¾‹</th></tr>
            """
            for col, miss in missing_top5.items():
                ratio = miss / len(df) * 100
                html_template += f"<tr><td>{col}</td><td>{miss:,}</td><td>{ratio:.1f}%</td></tr>\n"
            html_template += "</table>"

            # æ•°å€¼å‹ç»Ÿè®¡è¡¨
            if num_stats is not None:
                html_template += f"""
                    <h2>ğŸ”¢ æ•°å€¼å‹å˜é‡ - æè¿°æ€§ç»Ÿè®¡</h2>
                    {num_stats.to_html(classes='table', float_format='{:,.0f}'.format)}
                """

            # åˆ†ç±»å˜é‡ç¤ºä¾‹
            if cat_example:
                html_template += """
                    <h2>ğŸ·ï¸ åˆ†ç±»å˜é‡ - æœ€å¸¸è§å€¼ï¼ˆå‰3ä¸ªï¼‰</h2>
                    <table>
                        <tr><th>å­—æ®µå</th><th>æœ€å¸¸è§å€¼</th><th>ç¼ºå¤±æ•°é‡</th></tr>
                """
                for col, info in cat_example.items():
                    html_template += f"<tr><td>{col}</td><td>{info['most']}</td><td>{info['missing']:,}</td></tr>\n"
                html_template += "</table>"

            # æ ·æœ¬æ•°æ®
            html_template += f"""
                    <h2>ğŸ“‹ æ ·æœ¬æ•°æ®é¢„è§ˆï¼ˆè½¬ç½®ï¼‰</h2>
                    {sample_transposed.to_html(classes='table', float_format='{:,.0f}'.format)}
                </div>

                <footer style="margin-top: 50px; color: #777; text-align: center;">
                    ç”Ÿæˆå·¥å…·ï¼šPython + Pandas | æ•°æ®æ¢æŸ¥æŠ¥å‘Š
                </footer>
            </body>
            </html>
            """

            # ä¿å­˜æ–‡ä»¶
            with open(filename, 'w', encoding='utf-8') as f:
                f.write(html_template)

            print(f"\nâœ… HTML æŠ¥å‘Šå·²ä¿å­˜è‡³ï¼š{filename}")

    # å‡½æ•°ç»“æŸ


# =============================================================================
# è°ƒç”¨å‡½æ•°å¹¶å¯¼å‡º HTML æŠ¥å‘Š
# =============================================================================

# æ¢æŸ¥ NLS97 å¹¶ç”ŸæˆæŠ¥å‘Š
describe_dataset(nls97, "NLS97 æ•°æ®é›†", export_html=True)

# æ¢æŸ¥ Covid æ•°æ®å¹¶ç”ŸæˆæŠ¥å‘Š
describe_dataset(covidtotals, "Covid ç´¯è®¡ç—…ä¾‹æ•°æ®é›†", export_html=True)
```

```python
# -*- coding: utf-8 -*-
"""
AI-NLS97 æ•°æ®åˆ†ææ¼”ç¤ºï¼ˆå‡½æ•°åŒ–é‡æ„ç‰ˆï¼‰
==================================
é€šè¿‡å‡½æ•°å°è£…å®ç°ï¼š
- æ•°æ®åŠ è½½ä¸æ¸…æ´—
- åˆ—é€‰æ‹©ç­–ç•¥
- æŒ‰ä¸»é¢˜é‡æ’åˆ—é¡ºåº
- å¿«é€Ÿæ•°æ®æ£€æŸ¥
é€‚ç”¨äºæ•™å­¦æ¼”ç¤ºå’Œå¯å¤ç”¨åˆ†ææµç¨‹ã€‚
"""

import pandas as pd


# -----------------------------
# 1. è®¾ç½®æ˜¾ç¤ºé€‰é¡¹
# -----------------------------
def set_display_options():
    """è®¾ç½® Pandas æ˜¾ç¤ºé€‰é¡¹ï¼Œä¾¿äºæŸ¥çœ‹æ•°æ®"""
    pd.set_option('display.width', 78)
    pd.set_option('display.max_columns', 8)
    pd.set_option('display.max_rows', 15)
    pd.options.display.float_format = '{:,.0f}'.format


# -----------------------------
# 2. åŠ è½½ä¸åŸºç¡€é¢„å¤„ç†
# -----------------------------
def load_nls97(filepath="data/nls97.csv"):
    """åŠ è½½ NLS97 æ•°æ®å¹¶è®¾ç½®ç´¢å¼•"""
    df = pd.read_csv(filepath)
    df.set_index("personid", inplace=True)
    return df


def convert_object_to_category(df):
    """å°†æ‰€æœ‰ object ç±»å‹åˆ—è½¬æ¢ä¸º category"""
    df[df.select_dtypes(['object']).columns] = \
        df.select_dtypes(['object']).transform(lambda x: x.astype('category'))
    return df


# -----------------------------
# 3. åˆ—é€‰æ‹©å·¥å…·å‡½æ•°
# -----------------------------
def select_columns(df, cols, method='direct'):
    """
    å¤šç§æ–¹å¼é€‰æ‹©åˆ—ï¼ˆç”¨äºæ¼”ç¤ºä¸åŒè¯­æ³•ï¼‰
    """
    if method == 'direct':
        return df[cols]
    elif method == 'loc':
        return df.loc[:, cols]
    elif method == 'iloc':
        col_indices = [df.columns.get_loc(c) for c in cols]
        return df.iloc[:, col_indices]
    else:
        raise ValueError("method å¿…é¡»æ˜¯ 'direct', 'loc', æˆ– 'iloc'")


def filter_by_keyword(df, keyword):
    """ä½¿ç”¨ filter æ¨¡ç³ŠåŒ¹é…åˆ—å"""
    return df.filter(like=keyword)


def select_by_dtypes(df, include=None, exclude=None):
    """æ ¹æ®æ•°æ®ç±»å‹é€‰æ‹©åˆ—"""
    return df.select_dtypes(include=include, exclude=exclude)


# -----------------------------
# 4. æ„å»ºä¸»é¢˜å˜é‡ç»„
# -----------------------------
def get_variable_groups():
    """å®šä¹‰å„ä¸»é¢˜çš„å˜é‡ç»„"""
    return {
        'demoadult': [
            'highestgradecompleted', 'maritalstatus',
            'childathome', 'childnotathome', 'wageincome',
            'weeklyhrscomputer', 'weeklyhrstv', 'nightlyhrssleep',
            'highestdegree'
        ],
        'demo': ['gender', 'birthmonth', 'birthyear'],
        'highschoolrecord': [
            'satverbal', 'satmath', 'gpaoverall',
            'gpaenglish', 'gpamath', 'gpascience'
        ],
        'govresp': [
            'govprovidejobs', 'govpricecontrols', 'govhealthcare',
            'govelderliving', 'govindhelp', 'govunemp',
            'govincomediff', 'govcollegefinance',
            'govdecenthousing', 'govprotectenvironment'
        ],
        'weeksworked': [f"weeksworked{i:02d}" for i in range(0, 18)],
        'colenr': [f"colenr{suffix}{year:02d}"
                   for year in range(97, 18)  # 97 to 17
                   for suffix in ['feb', 'oct']]
    }


def reorder_columns(df, groups):
    """
    æŒ‰ç…§æŒ‡å®šåˆ†ç»„é¡ºåºé‡æ–°æ’åˆ—åˆ—
    åªä¿ç•™åŸå§‹æ•°æ®ä¸­å­˜åœ¨çš„åˆ—
    """
    new_order = []
    for group_name, columns in groups.items():
        valid_cols = [col for col in columns if col in df.columns]
        new_order.extend(valid_cols)
        print(f"âœ… {group_name}: {len(valid_cols)} åˆ—")
    return df[new_order]


# -----------------------------
# 5. æ¼”ç¤ºä¸æ£€æŸ¥å‡½æ•°
# -----------------------------
def show_selection_demo(df):
    """æ¼”ç¤ºä¸åŒå•åˆ—é€‰æ‹©æ–¹å¼çš„è¿”å›ç±»å‹"""
    print("=== å•åˆ—é€‰æ‹©ï¼šè¿”å›ç±»å‹å¯¹æ¯” ===")
    examples = [
        ("ç›´æ¥ç´¢å¼•", df['gender'], type(df['gender']).__name__),
        ("åŒæ‹¬å·", df[['gender']], type(df[['gender']]).__name__),
        ("loc", df.loc[:, ['gender']], type(df.loc[:, ['gender']]).__name__),
        ("iloc", df.iloc[:, [0]], type(df.iloc[:, [0]]).__name__)
    ]
    for desc, obj, typ in examples:
        print(f"{desc:8} â†’ {typ}")


def quick_inspect(df):
    """å¿«é€ŸæŸ¥çœ‹æ•°æ®çŠ¶æ€"""
    print(f"\nğŸ“Š æ•°æ®å½¢çŠ¶: {df.shape}")
    print(f"ğŸ“ éç±»åˆ«å‹å˜é‡:")
    num_df = select_by_dtypes(df, exclude=["category"])
    num_df.info(verbose=False, show_counts=True)

    print(f"\nğŸ” åŒ…å« 'income' çš„åˆ—:")
    income_cols = filter_by_keyword(df, 'income')
    print(income_cols.columns.tolist())
    print(income_cols.head(3))


# -----------------------------
# 6. ä¸»æµç¨‹å‡½æ•°
# -----------------------------
def main():
    """ä¸»å‡½æ•°ï¼šæ‰§è¡Œå®Œæ•´çš„æ•°æ®åŠ è½½ â†’ æ¸…æ´— â†’ é‡ç»„ â†’ æ£€æŸ¥æµç¨‹"""
    print("ğŸš€ å¼€å§‹ NLS97 æ•°æ®åˆ†ææ¼”ç¤º...\n")

    # æ­¥éª¤ 1: è®¾ç½®æ˜¾ç¤º
    set_display_options()

    # æ­¥éª¤ 2: åŠ è½½æ•°æ®
    print("ğŸ“ æ­£åœ¨åŠ è½½æ•°æ®...")
    df = load_nls97()
    print(f"åŸå§‹åˆ—æ•°: {len(df.columns)}\n")

    # æ­¥éª¤ 3: ç±»å‹è½¬æ¢
    print("ğŸ”§ è½¬æ¢æ–‡æœ¬åˆ—ä¸º category...")
    df = convert_object_to_category(df)

    # æ­¥éª¤ 4: æ¼”ç¤ºåˆ—é€‰æ‹©æ–¹æ³•
    show_selection_demo(df)

    # æ­¥éª¤ 5: æ„å»ºå˜é‡ç»„å¹¶é‡æ’åº
    print("\nğŸ”„ æ­£åœ¨æŒ‰ä¸»é¢˜æ•´ç†åˆ—é¡ºåº...")
    var_groups = get_variable_groups()
    df = reorder_columns(df, var_groups)
    print(f"âœ… æœ€ç»ˆåˆ—æ•°: {df.shape[1]}")

    # æ­¥éª¤ 6: å¿«é€Ÿæ£€æŸ¥
    quick_inspect(df)

    # å¯é€‰ï¼šä¿å­˜æ¸…æ´—åæ•°æ®
    output_dir = "results"
    os.makedirs(output_dir, exist_ok=True)
    df.to_csv(os.path.join(output_dir, "nls97_cleaned.csv"), index=True)
    print(f"\nğŸ‰ æ•°æ®å¤„ç†å®Œæˆï¼Œå¹¶å·²ä¿å­˜è‡³ '{output_dir}/nls97_cleaned.csv'")


# -----------------------------
# è¿è¡Œä¸»ç¨‹åº
# -----------------------------
if __name__ == "__main__":
    main()
```
